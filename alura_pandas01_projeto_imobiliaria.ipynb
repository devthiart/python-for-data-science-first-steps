# Anális exploratória de dados

## Qual o valor médio de aluguel por tipo de imóvel?

```python
data.head()
```

```python
data['Valor'].mean() # Average of values ('Valor')
```

```python
data.groupby('Tipo').mean(numeric_only=True)
```

```python
data.groupby('Tipo')['Valor'].mean()
```

```python
data.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
df_value_type = data.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_value_type.plot(kind='barh', figsize=(14, 10), color='purple')
```
